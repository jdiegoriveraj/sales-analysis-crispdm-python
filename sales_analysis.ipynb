```python
# ============================================================
# SALES ANALYSIS USING CRISP-DM METHODOLOGY
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# Visualization configuration
# -----------------------------
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

# -----------------------------
# Load data
# -----------------------------
df = pd.read_excel("bcleansales2025.xlsx")

print("=" * 60)
print("CRISP-DM PHASE: DATA UNDERSTANDING")
print("=" * 60)

# Preview data
print("\n First 5 transactions:")
display(df.head())

# Dataset structure
print("\n Dataset info:")
display(df.info())

# Descriptive statistics
print("\n Descriptive statistics:")
display(df.describe())

# ============================================================
# CRISP-DM PHASE: DATA PREPARATION
# ============================================================

print("\n" + "=" * 60)
print("CRISP-DM PHASE: DATA PREPARATION")
print("=" * 60)

# Convert date column
df['Sale Date'] = pd.to_datetime(df['Sale Date'], format='%d/%m/%Y')

# Extract date components
df['Year'] = df['Sale Date'].dt.year
df['Month'] = df['Sale Date'].dt.month
df['Month_Name'] = df['Sale Date'].dt.month_name()
df['Weekday'] = df['Sale Date'].dt.day_name()

# Check missing values
print("\n Missing values by column:")
display(df.isnull().sum())

# Check duplicates
print(f"\n Duplicate records: {df.duplicated().sum()}")

# Dataset summary
print("\n Clean dataset summary:")
print(f"Transactions: {len(df)}")
print(f"Period: {df['Sale Date'].min().date()} â†’ {df['Sale Date'].max().date()}")
print(f"Unique customers: {df['Customer'].nunique()}")
print(f"Unique products: {df['Product'].nunique()}")
print(f"Total sales: ${df['Total Sale'].sum():,.0f}")

# ============================================================
# CRISP-DM PHASE: MODELING â€“ DESCRIPTIVE ANALYSIS
# ============================================================

print("\n" + "=" * 60)
print("CRISP-DM PHASE: MODELING â€“ DESCRIPTIVE ANALYSIS")
print("=" * 60)

# KPIs
total_sales = df['Total Sale'].sum()
num_transactions = len(df)
avg_ticket = df['Total Sale'].mean()
num_customers = df['Customer'].nunique()

print("\n KEY BUSINESS KPIs")
print(f"ðŸ’° Total Sales: ${total_sales:,.0f}")
print(f"ðŸ§¾ Transactions: {num_transactions}")
print(f"ðŸ’³ Average Ticket: ${avg_ticket:,.0f}")
print(f"ðŸ‘¥ Unique Customers: {num_customers}")
print(f"ðŸ›’ Avg Purchases per Customer: {num_transactions / num_customers:.1f}")

# Monthly sales
monthly_sales = (
    df.groupby('Month_Name')['Total Sale']
    .sum()
    .reindex([
        'February', 'March', 'April', 'May', 'June',
        'July', 'August', 'September', 'October', 'November'
    ])
)

print("\n MONTHLY SALES:")
for month, sales in monthly_sales.items():
    print(f"{month}: ${sales:,.0f}")

# ============================================================
# PRODUCT ANALYSIS + ABC / PARETO
# ============================================================

print("\n" + "=" * 60)
print("CRISP-DM PHASE: MODELING â€“ PRODUCT ANALYSIS")
print("=" * 60)

products = df.groupby('Product').agg({
    'ID Sale': 'count',
    'Amount': 'sum',
    'Total Sale': 'sum'
}).reset_index()

products.columns = ['Product', 'Num_Sales', 'Units_Sold', 'Revenue']
products = products.sort_values('Revenue', ascending=False)

products['Percentage'] = (products['Revenue'] / products['Revenue'].sum()) * 100
products['Cumulative_Percentage'] = products['Percentage'].cumsum()

products['ABC_Category'] = products['Cumulative_Percentage'].apply(
    lambda x: 'A' if x <= 80 else ('B' if x <= 95 else 'C')
)

print("\n TOP 10 PRODUCTS BY REVENUE:")
display(products.head(10))

# ============================================================
# CRISP-DM PHASE: MODELING â€“ RFM SEGMENTATION
# ============================================================

print("\n" + "=" * 60)
print("CRISP-DM PHASE: MODELING â€“ RFM SEGMENTATION")
print("=" * 60)

reference_date = df['Sale Date'].max() + pd.Timedelta(days=1)

rfm = df.groupby('Customer').agg({
    'Sale Date': lambda x: (reference_date - x.max()).days,
    'ID Sale': 'count',
    'Total Sale': 'sum'
}).reset_index()

rfm.columns = ['Customer', 'Recency', 'Frequency', 'Monetary']

# RFM scoring function
def rfm_score(series, ascending=True):
    try:
        return pd.qcut(series, 5, labels=[5,4,3,2,1] if ascending else [1,2,3,4,5], duplicates='drop')
    except ValueError:
        return pd.cut(series, 5, labels=[5,4,3,2,1] if ascending else [1,2,3,4,5])

rfm['R_Score'] = rfm_score(rfm['Recency'], ascending=True)
rfm['F_Score'] = rfm_score(rfm['Frequency'].rank(method='first'), ascending=False)
rfm['M_Score'] = rfm_score(rfm['Monetary'].rank(method='first'), ascending=False)

rfm[['R_Score','F_Score','M_Score']] = rfm[['R_Score','F_Score','M_Score']].fillna(3)

# Customer segmentation
def segment_customer(row):
    r, f, m = int(row['R_Score']), int(row['F_Score']), int(row['M_Score'])
    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'
    elif f >= 4 and m >= 4:
        return 'Loyal'
    elif r >= 4 and f <= 2:
        return 'Potential'
    elif r <= 2 and f >= 3:
        return 'At Risk'
    elif r <= 2 and f <= 2:
        return 'Lost'
    else:
        return 'Occasional'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)

rfm_sorted = rfm.sort_values('Monetary', ascending=False)

# Save RFM
rfm.to_csv("rfm_analysis.csv", index=False)

# ============================================================
# CRISP-DM PHASE: DEPLOYMENT â€“ VISUALIZATIONS
# ============================================================

print("\n" + "=" * 60)
print("CRISP-DM PHASE: DEPLOYMENT â€“ VISUALIZATIONS")
print("=" * 60)

# 1. Monthly Sales Trend
monthly_sales.plot(marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("1_monthly_sales_trend.png", dpi=300)
plt.show()

# 2. Top 10 Customers
top_customers = rfm_sorted.head(10)
plt.barh(top_customers['Customer'], top_customers['Monetary'])
plt.title("Top 10 Customers by Revenue")
plt.xlabel("Revenue ($)")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig("2_top_customers.png", dpi=300)
plt.show()

# 3. RFM Segments Distribution
rfm['Segment'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("RFM Segments Distribution")
plt.ylabel("")
plt.tight_layout()
plt.savefig("3_rfm_segments_distribution.png", dpi=300)
plt.show()

# 4. Top Products
plt.barh(products.head(10)['Product'], products.head(10)['Revenue'])
plt.title("Top 10 Products by Revenue")
plt.xlabel("Revenue ($)")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig("4_top_products.png", dpi=300)
plt.show()

# 5. Pareto Products
plt.bar(products['Product'], products['Revenue'])
plt.plot(products['Product'], products['Cumulative_Percentage'], color='red', marker='o')
plt.axhline(80, color='green', linestyle='--')
plt.xticks([])
plt.title("Pareto Analysis â€“ Products (80/20 Rule)")
plt.ylabel("Revenue / Cumulative %")
plt.tight_layout()
plt.savefig("5_pareto_products.png", dpi=300)
plt.show()

print("\n" + "=" * 60)
print("CRISP-DM PHASE: EVALUATION & CONCLUSIONS")
print("=" * 60)

print("\n KEY FINDINGS:")

# -----------------------------
# 1. Customers
# -----------------------------
print("\n1. CUSTOMERS:")
top_20_customers = int(len(rfm) * 0.2)

print(
    f"   - Top 20% of customers ({top_20_customers}) generate "
    f"${rfm_sorted.head(top_20_customers)['Monetary'].sum():,.0f} in revenue"
)

one_time_customers = rfm[rfm['Frequency'] == 1]
print(
    f"   - {len(one_time_customers)} customers purchased only once "
    f"({len(one_time_customers) / len(rfm) * 100:.1f}%)"
)

print(
    f"   - Champions segment size: "
    f"{len(rfm[rfm['Segment'] == 'Champions'])} customers"
)

# -----------------------------
# 2. Products
# -----------------------------
print("\n2. PRODUCTS:")
print(
    f"   - Top 3 products generate: "
    f"${products.head(3)['Revenue'].sum():,.0f}"
)

print(
    f"   - #1 Product: {products.iloc[0]['Product']} "
    f"with ${products.iloc[0]['Revenue']:,.0f}"
)

print(
    f"   - Category A products (80% of sales): "
    f"{len(products[products['ABC_Category'] == 'A'])}"
)

# -----------------------------
# 3. Seasonality
# -----------------------------
print("\n3. SEASONALITY:")
best_month = monthly_sales.idxmax()
worst_month = monthly_sales.idxmin()

print(
    f"   - Best month: {best_month} "
    f"(${monthly_sales[best_month]:,.0f})"
)

print(
    f"   - Lowest sales month: {worst_month} "
    f"(${monthly_sales[worst_month]:,.0f})"
)

# ============================================================
# Strategic Recommendations
# ============================================================

print("\n STRATEGIC RECOMMENDATIONS:")

print("\n1. CHAMPIONS CUSTOMER RETENTION:")
print("   - Create an exclusive loyalty program")
print("   - Offer volume-based discounts")
print("   - Monthly personalized follow-ups")

print("\n2. AT-RISK CUSTOMER RECOVERY:")
print("   - Identify churn reasons")
print("   - Launch reactivation campaigns with special offers")
print("   - Post-sale follow-up and feedback collection")

print("\n3. INVENTORY OPTIMIZATION:")
print("   - Prioritize stock for Category A products")
print("   - Evaluate discontinuing low-rotation Category C products")
print("   - Negotiate better pricing with top-product suppliers")

print("\n4. COMMERCIAL STRATEGY:")
print("   - Focus on flagship products")
print("   - Promote bundles for Potential customers")
print("   - Implement cross-selling strategies")

print("\n FULL ANALYSIS COMPLETED SUCCESSFULLY")
print("=" * 60)
```
